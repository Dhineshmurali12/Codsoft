```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
train_path = "/content/train_data.csv"
train_data = pd.read_csv(train_path, sep = ':::', names = ['Title', 'Genre', 'Description'], engine = 'python')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_20newsgroups

# Assuming 'text' and 'genre' are the column names in your DataFrame
text_column = 'Title'  # Replace 'Title' with the actual column name containing the text data
genre_column = 'Genre'  # Replace 'Genre' with the actual column name containing the labels

# Load dataset
X_train, X_test, y_train, y_test = train_test_split(train_data[text_column], train_data[genre_column], test_size=0.2, random_state=42)

# Build pipeline with TF-IDF and Logistic Regression
model = make_pipeline(TfidfVectorizer(), LogisticRegression(max_iter=1000))
model.fit(X_train, y_train)

# Function to predict the genre
def predict_genre(plot_summary):
    prediction = model.predict([plot_summary])
    # Get the index of the predicted class
    # Convert prediction to numerical index if it's not already
    predicted_index = prediction[0]
    if isinstance(predicted_index, str):
        unique_genres = train_data[genre_column].unique()
        predicted_index = np.where(unique_genres == predicted_index)[0][0]  # Find index of predicted genre

    # Get the list of unique genres
    unique_genres = train_data[genre_column].unique()
    # Check if the predicted index is within the bounds of the unique genres array
    if 0 <= predicted_index < len(unique_genres):
        return unique_genres[predicted_index]  # Use unique values from the genre column
    else:
        return "Unknown Genre" # Handle the case where the prediction is out of bounds

# Get user input and predict genre
user_input = input("Enter a movie plot summary: ")
print("Predicted Genre:", predict_genre(user_input))
```
